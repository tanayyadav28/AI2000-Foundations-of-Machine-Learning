# Q5

```python
#importing the required libraries

import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score
import csv
```

```python
# creating the dataframes and processing the data for fitting

train_data = pd.read_csv('gisette_train.csv', header = None)
test_data = pd.read_csv('gisette_valid.csv', header = None)
train_data = train_data.drop(columns = 5000)
test_data = test_data.drop(columns = 5000)
```

```python
# displaying the training dataframe

train_data
```

```python
# displaying the testing dataframe

test_data
```

```python
# creating the arrays for training and testing labels using loadtxt function from numpy

with open('gisette_train_labels.csv') as f:
    train_label = np.loadtxt(f, delimiter = ',')
    
with open('gisette_valid_labels.csv') as f:
    test_label = np.loadtxt(f, delimiter = ',')
```

```python
# verifying the shape of the array containing training labels

train_label.shape
```

```python
# verifying the shape of the array containing testing labels

test_label.shape
```

# Q5(a)

```python
# creating the classifier clf using the sklearn.svm library
# linear kernel 

clf = svm.SVC(kernel = 'linear')
```

```python
# fitting the classifier using the training dataset

clf.fit(train_data, train_label)
```

```python
# predicting the labels on the training dataset

y_train = clf.predict(train_data)

#calculating the accuracy of the predictions over the training dataset

train_accu = accuracy_score(train_label, y_train)
print('Training error =', 1 - train_accu)
```

```python
# predicting the labels over the testing dataset

y_test = clf.predict(test_data)

#calculating the accuracy of the predictions over the testing dataset

test_accu = accuracy_score(test_label, y_test)
print('Testing error =', 1 - test_accu)
```

```python
# finding the number of support vectors for both the classes

vectors = clf.n_support_
print('Number of support vectors =', vectors)
```

# Q5(b)

Gaussian Kernel with $\gamma = 0.001$

```python
# creating the classifier clf using the sklearn.svm library
# rbf kernel

clf_rbf = svm.SVC(kernel = 'rbf', gamma = 0.0001)

# fitting the classifier using the training dataset
clf_rbf.fit(train_data, train_label)

# predicting the labels on the training dataset
y_rbf_train = clf_rbf.predict(train_data)

#calculating the accuracy of the predictions over the training dataset
rbf_train_accu = accuracy_score(train_label, y_rbf_train)

# predicting the labels over the testing dataset
y_rbf_test = clf_rbf.predict(test_data)

#calculating the accuracy of the predictions over the testing dataset
rbf_test_accu = accuracy_score(test_label, y_rbf_test)

# finding the number of support vectors for both the classes
rbf_vectors = clf_rbf.n_support_

print('For RBF Kernel:')
print('Training error =', 1 - rbf_train_accu)
print('Testing error =', 1 - rbf_test_accu)
print('Number of support vectors =', rbf_vectors)
```

Polynomial Kernel with $C = 1$ and $Q = 2$ i.e. $(1 + x^Tx)^2$

```python
# creating the classifier clf using the sklearn.svm library
# polynomial kernel

clf_poly = svm.SVC(kernel = 'poly', coef0 = 1, degree = 2)

# fitting the classifier using the training dataset
clf_poly.fit(train_data, train_label)

# predicting the labels on the training dataset
y_poly_train = clf_poly.predict(train_data)

#calculating the accuracy of the predictions over the training dataset
poly_train_accu = accuracy_score(train_label, y_poly_train)

# predicting the labels over the testing dataset
y_poly_test = clf_poly.predict(test_data)

#calculating the accuracy of the predictions over the testing dataset
poly_test_accu = accuracy_score(test_label, y_poly_test)

# finding the number of support vectors for both the classes
poly_vectors = clf_poly.n_support_

print('For Polynomial Kernel:')
print('Training error =', 1 - poly_train_accu)
print('Testing error =', 1 - poly_test_accu)
print('Number of support vectors =', poly_vectors)
```
